#### Introduction of DataSet (Most_Streamed_Spotify_Songs_2024) : 

This dataset comprises the most played songs on Spotify, encompassing a wide array of metrics that capture the songs' performance and engagement across several platforms. The data includes detailed information such as the track's name, album, artist, release date, and various performance metrics. Specifically, the dataset tracks:

**Spotify Metrics**: Streams, playlist count, playlist reach, and popularity.
**YouTube Metrics**: Views and likes.
**TikTok Metrics**: Posts, likes, and views.
**Apple Music Metrics**: Playlist count.
**Additional Metrics**: AirPlay spins, Amazon playlist count, and Shazam counts.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
pd.set_option('display.max_columns', None)
```

#### Reading csv file and creating a dataframe :
1. The data was not 'UTF-8' encoded hence 'latin1' encoding is used.
2. The type of each row was object, hence the columns containing numeric values were converted into float.
3. The numeric data had commas in it (for example, 1,34,789) hence ',' was converted to ''.

```python
df = pd.read_csv('Most_Streamed_Spotify_Songs_2024.csv', encoding='latin1')

columns_to_convert = [
    'All Time Rank',
    'Track Score',
    'Spotify Streams',
    'Spotify Playlist Count',
    'Spotify Playlist Reach',
    'Spotify Popularity',
    'YouTube Views',
    'YouTube Likes',
    'TikTok Posts',
    'TikTok Likes',
    'TikTok Views',
    'YouTube Playlist Reach',
    'Apple Music Playlist Count',
    'AirPlay Spins',
    'SiriusXM Spins',
    'Deezer Playlist Count',
    'Deezer Playlist Reach',
    'Amazon Playlist Count',
    'Pandora Streams',
    'Pandora Track Stations',
    'Soundcloud Streams',
    'Shazam Counts',
    'Explicit Track'
]
for column in columns_to_convert:
    df[column] = df[column].astype(str).str.replace(',', '').astype(float)
```

```python
# Define a function to check if a string contains only ASCII characters
def is_ascii(s):
    if isinstance(s, str):
        return all(ord(c) < 128 for c in s)
    return False

# Remove rows where 'Track' or 'Album Name' contain non-ASCII characters
df_cleaned = df[df['Track'].apply(is_ascii) & df['Album Name'].apply(is_ascii) & df['Artist'].apply(is_ascii)]

# Save the cleaned DataFrame back to the original CSV file (Optional)
#df_cleaned.to_csv('Most_Streamed_Spotify_Songs_2024.csv', index=False)
```

```python
def fill_value_with_mean(column):
    df_cleaned[column].fillna(df_cleaned[column].mean(), inplace=True)

# List of columns to be filled with mean value
columns_to_fill = [
    'Spotify Streams',
    'Spotify Playlist Count',
    'Spotify Playlist Reach',
    'Spotify Popularity',
    'YouTube Views',
    'YouTube Likes',
    'TikTok Posts',
    'TikTok Likes',
    'TikTok Views',
    'YouTube Playlist Reach',
    'Apple Music Playlist Count',
    'AirPlay Spins',
    'Amazon Playlist Count',
    'Shazam Counts'
]

for column in columns_to_fill:
    fill_value_with_mean(column)
```

```python
# Define the columns corresponding to different platforms
platform_columns = {
    'Spotify': ['Spotify Playlist Count', 'Spotify Playlist Reach', 'Spotify Streams'],
    'YouTube': ['YouTube Views', 'YouTube Likes', 'YouTube Playlist Reach'],
    'TikTok': ['TikTok Posts', 'TikTok Likes', 'TikTok Views']
}

# Sum the values for each platform
platform_sums = {}
for platform, columns in platform_columns.items():
    platform_sums[platform] = df[columns].sum().sum()

# Extract data for plotting
platforms = list(platform_sums.keys())
counts = list(platform_sums.values())

# Create a pie chart
plt.figure(figsize=(10, 6))
plt.pie(counts, labels=platforms, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribution of Playlist Counts, Likes, Views, and Posts Across Platforms')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
```

#### Observation
This pie chart show sthe distribution of playlist counts, likes, views, and posts across different platforms.

The chart provides a visual representation of how these interactions and counts are distributed across different platforms. For instance, it helps to see that TikTok has more interactions compared to YouTube or Spotify in terms of playlist counts, likes, views, and posts.
